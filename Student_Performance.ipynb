## **Heatmap Analysis**

Our next step is to check either the educational level of parent effect the student score, for this purpose we use heatmap beacuse heatmap elaborate it clearly.
Firstly we use group by fuction to grouped the 'math score', 'reading score' and 'writing score' and then we use print function to print the GB(grouped by) data.


```python
GB= df.groupby('parental level of education').agg({'math score':'mean', 'reading score':'mean', 'writing score':'mean'})
print(GB)
```

```python
sns.heatmap(GB, annot=True)
plt.title("Relationship between Parent's Education & Students scores" )
plt.show()
```

Now from the above graph we can seen that the Parent's Education has a good impact on a Students score. We can clearly observe that the students whose Parents have done Master's degree score higher compare to others.

## **Distribution of Race/Ethnicity Group using PIE Chart**

```python
groupA= df.loc[df['race/ethnicity'] == 'group A'].count()
groupB= df.loc[df['race/ethnicity'] == 'group B'].count()
groupC= df.loc[df['race/ethnicity'] == 'group C'].count()
groupD= df.loc[df['race/ethnicity'] == 'group D'].count()
groupE= df.loc[df['race/ethnicity'] == 'group E'].count()

l=["group A","group B","group C","group D","group E"]
mylist=[groupA["race/ethnicity"],groupB["race/ethnicity"],groupC["race/ethnicity"],groupD["race/ethnicity"],groupE["race/ethnicity"]]
plt.pie(mylist, labels=l, autopct='%1.1f%%')
plt.title('Distribution of Race/Ethnicity Group')

```
